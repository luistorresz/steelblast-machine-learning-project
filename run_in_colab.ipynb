```python
# Remove previous cloned repository if it exists to ensure a clean clone
import os
import shutil

repo_name = 'steelblast-machine-learning-project'
if os.path.exists(repo_name):
    shutil.rmtree(repo_name)
    print(f"Removed existing directory: {repo_name}")
else:
    print(f"Directory {repo_name} does not exist. Proceeding with clone.")

!git clone https://github.com/luistorresz/steelblast-machine-learning-project.git    

import sys
import os


module_subfolder = 'modules' # Set to empty string as the module is directly in the repo_name

# Construct the full path to the directory containing the module
full_path = os.path.join(repo_name, module_subfolder) if module_subfolder else repo_name

if os.path.exists(full_path) and full_path not in sys.path:
    sys.path.insert(0, full_path)
    print(f"Added '{full_path}' to sys.path.")
else:
    print(f"Directory '{full_path}' not found or already in sys.path. Please ensure the repository was cloned correctly and the path is correct.")

import requests
import zipfile
import io
def download_and_extract_zip(extract_to: str):
    """
    Download a zip file from a DATAVERSE and extract its contents to a specified directory.
    If the specified directory does not exist, it will be created.

    Args:
        extract_to (str): The directory to extract the contents of the zip file to.
    """
    dataverse_url = "https://dataverse.nl/api/access/dataset/:persistentId/?persistentId=doi:10.34894/EKZNN0"

    response = requests.get(dataverse_url)

    if response.status_code == 200:
        if 'application/zip' not in response.headers.get('Content-Type', ''):
            print("The downloaded file is not a zip file.")
            print("Please try to download the file manually from the Dataverse.")
            print("See the ReadMe for more information.")
            return
        with zipfile.ZipFile(io.BytesIO(response.content)) as z:
            z.extractall(extract_to)
            print(f"Extracted files to: {extract_to}")

    else:
        print(f"Failed to download file: {response.status_code}")
        print("Please try to download the file manually from the Dataverse.")
        print("See the ReadMe for more information.")
download_and_extract_zip(".")

%run "/content/steelblast-machine-learning-project/steelblast_transfer_learning.ipynb"

```
